## Craft your materials
- Write the narrative, sell the narrative
- Be like Rita Skeeter when applying to jobs. 
    - Be concise and grab attention
    
## Resume Pro-Tips
- Keep it short
- Use clear words
- Use industry standard terms
- Enlist a good editor/Linda
- Pare daown for relevance
- Consider visual real estate
- If contacted by a recruiter, ask what appeals to you, what draws people to you. Identify a top strength


## Apply
- Even perfectly qualified candidates really only get contacted 10-15% of the time. 
    - Even the best rarely get callbacks (this is why networking really comes in to help)
- Resumes should be tailored for the job

## Rock the interview
- You are allowed to say "um"
- Hiring committees want someone who is a team player, enjoys problem solving, can accept feedback, can work with for years

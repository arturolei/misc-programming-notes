## Be a Good Reviewer:
- Code review should not be a trip to the therapist
- Have empathy
- State of mind going into a review
- It's the code not the person
- Asks questions
- Offer suggestions not ultimatums
- Avoid these terms:
    - Simply
    - Easily
    - Just
    - Obviously
    - Well, actually
- Share the how and why 
    - Feel free to back-up with links to stack overflow
- Remember to praise the positive; it's not all negative
- Perfect workable code, not just code that suits you. 
- Do not burn out
    - Review 200-400 lines; this is a standard limit
    - Post 500 lines of code at once makes things not make sense; reduce effectiveness at looking at code
- Do review code reviews within 24-48 hours
- See Recurse Center Social Rules

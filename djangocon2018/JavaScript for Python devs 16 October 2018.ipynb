## JavaScript Versions
- JavaScript is language standardized by ECMA standard
- The oldest version supported by older browsers is ES5 (the one you know)
- ES6 or ECMAScript 2015 is the newest version (shift to a yearly release model?)
    - Supported by modern browsers but not older browsers
    - Following this yearly release

## Syntax:
- More parentheses
- Weakly typed, dynamic ("1" +2
- Var, Let, Const
- Let, Const for scope reasons

## Scope
- Every JS program can define or modify the global scope
    - Okay in smaller program, but dangerous in larger scope
    - Do not mess with global scope
    
## DataTypes
- Data has five or so primitive/simple types
    - Everything else is an object
- Objects are mutable
    -Key to JS being a functional language

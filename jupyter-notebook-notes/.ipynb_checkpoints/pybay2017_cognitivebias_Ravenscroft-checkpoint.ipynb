"Are you really a programmer?"
- The questions we ask shape or reveal our perspectives. 
- What makes a programmer? A rocket scientist who codes might not be a programmer. 

Confirmation Bias Testing:
- You test for things that work but you really ought to test for things that do not work as well (edge cases and error-handling)

Case Study: 
- Blind auditions for an orchestra; visual bias removed but not all bias givenproblem: women wear heals)

Overcoming Stereotypes and Biases:
- Make concerted efforts to reach out to margianlized groups
- Actively work to bring in margianalized programmers and Pythonistas
- Encourage people not like you to actively challenge the patterns we have held up

Hiring:
- Avoid "gendered" ads
- Use "blind" resumes
    - remove name
    - remove university and other "clues" or indications of group membership
    
Interviewing candidates of marginalized groups:
- Look first for reasons to hire 
- Look for strengths and abilities
- Look for imposter syndrome 
    - Key phrases: "As part of a team"

Seek out new lifeforms who are not like you. 

At a technical conference, presume everyone is a programmer. 
- Do not ask people "Whether they are programmers?" but ask "What kind of programming? How do you use"
